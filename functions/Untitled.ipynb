```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def name_by_number():
    """
    Данная функция реализовывает поиск имени по номеру документа
    """

    num_doc = input('Введите номер документа: ')
    for doc in documents:
        if doc['number'] == str(num_doc):
            return(print(f'Владелец документа: {doc["name"]}'))
        else:
            continue
    return(print('Документ не найден в базе'))

def dir_by_number():
    """
    Данная функция возвращает номер директории, в которой расположен данный документ
    """ 
    
    num_doc = input('Введите номер документа: ')
    for direct in directories.items():
        if num_doc in direct[1]:
            return(print(f'Документ хранится на полке: {direct[0]}'))
        else:
            continue
    return(print('Документ не найден на полке'))



def info_on_doc():
    """
    Данная функция возвращает всю информацию по существующим документам
    """
    
    for i in documents:
        for k in directories.items():
            if i['number'] in k[1]:
                print(f'№: {i["number"]}, тип: {i["type"]}, владелец: {i["name"]}, полка хранения: {k[0]} ')
            else:
                continue    
    return None
  

    
def add_direct():
    """
    Данная функция добавляет полку
    """
    
    num_dir = str(input('Введите номер полки: '))
    if num_dir in directories.keys():
        return(print(f'Такая полка уже существует. Текущий перечень полок: {", ".join(map(str,directories.keys()))}.'))
    else:
        directories.setdefault(num_dir,[])
        return(print(f'Полка добавлена. Текущий перечень полок: {", ".join(map(str,directories.keys()))}.'))
    
def del_dir():
    """
    Данная функция удаляет полку
    """
    
    num_dir = str(input('Введите номер полки: '))
    if num_dir not in directories.keys():
        return(print(f'Такой полки не существует. Текущий перечень полок: {", ".join(map(str,directories.keys()))}.'))
    else:
        if len(directories[num_dir]) == 0:
            directories.pop(num_dir)
            return(print(f'Полка удалена. Текущий перечень полок: {", ".join(map(str,directories.keys()))}.'))
        else:
            return(print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {", ".join(map(str,directories.keys()))}.'))
        
def add_doc():
    """
    Данная функция добавляет новый документ
    """
    
    num_doc = str(input('Введите номер документа: '))
    type_doc = str(input('Введите тип документа: '))
    owner_doc = str(input('Введите владельца документа: '))
    num_dir = str(input('Введите полку для хранения: '))
    if num_dir not in directories.keys():
        print(f'Такой полки не существует. Добавьте полку командой as.')
        print('Текущий список документов:')
        return(info_on_doc())
    else:
        documents.append({'type': type_doc, 'number': num_doc, 'name': owner_doc})
        directories[num_dir].append(num_doc)
        print('Документ добавлен. Текущий список документов:')
        return(info_on_doc())
    

def del_doc():
    """
    Данная функция удаляет документ
    """
    
    num_doc = str(input('Введите номер документа: '))
    documents_copy = documents.copy()
    for i in documents_copy:
        if i['number'] == num_doc:
            documents.remove(i)
    
    for value in directories.values():
        if num_doc in value:
            value.remove(num_doc)
            print('Документ удален.')
            print('Текущий список документов:')
            return(info_on_doc())
    print('Документ не найден в базе.')
    print('Текущий список документов:')
    return(info_on_doc())
            
def move_doc():
    """
    Данная функция перемещает документ из полки в полку
    """

    num_doc = str(input('Введите номер документа: '))
    num_dir = str(input('Введите номер полки: '))   

    if num_dir not in directories.keys():
        return(print(f'Такой полки не существует. Текущий перечень полок: {", ".join(map(str,directories.keys()))}.'))
  
    num_docs = []
    for doc in documents:
        num_docs.append(doc['number'])
    if num_doc not in num_docs:
        print('Документ не найден в базе.')
        print('Текущий список документов:')
        return(info_on_doc())
    
    for value in directories.values():
        if num_doc in value:
            value.remove(num_doc)
    directories[num_dir].append(num_doc)
    print('Документ перемещен. Текущий список документов:')
    return(info_on_doc())
    
def general_func():
    """
    Основной механизм запуска данной консоли
    """
    
    command = ''
    while command != 'q':
        command = input('Введите команду: ')
        if command == 'p':
            name_by_number()
        elif command == 's':
            dir_by_number()
        elif command == 'l':
            info_on_doc()
        elif command == 'ads':
            add_direct()
        elif command == 'ds':
            del_dir()
        elif command == 'ad':
            add_doc()
        elif command == 'd':
            del_doc()
        elif command == 'm':
            move_doc()
        elif command == 'q':
            pass
        else:
            print('Некорректная команда')

            
        
    print('Всего доброго!')
        
```
