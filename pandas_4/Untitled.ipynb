#### Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

* если источник traffic_source равен yandex или google, то в source_type ставится organic
* для источников paid и email из России - ставим ad
* для источников paid и email не из России - ставим other
* все остальные варианты берем из traffic_source без изменений

```python
pd.read_csv('sales_db.csv', sep=',').head()
```

```python
df_log = pd.read_csv('visit_log.csv', sep = ';')
df_log.head()
```

```python
df_log['traffic_source'].unique()
```

```python
def source_type(df):
    if df['traffic_source'] == 'yandex' or df['traffic_source'] == 'google':
        return 'organic'
    elif df['traffic_source'] == 'paid' and df['region'] == 'Russia':
        return 'ad'
    elif df['traffic_source'] == 'paid' and df['region'] != 'Russia':
        return 'other'
    else:
        return df['traffic_source']
```

```python
df_log['source_type'] = df_log.apply(source_type, axis = 1)
```

#### Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
df_ratings_movies = pd.read_csv('ml-latest-small/ratings.csv')
df_ratings_movies.head()
```

```python
df_high_value_rating = pd.DataFrame(df_ratings_movies.groupby('userId').count()['rating']> 100).reset_index()
df_high_value_rating
```

```python
df_high_value_rating.rename(columns= {'rating': 'rating_over_100'}, inplace= True)
df_high_value_rating
```

```python
df_ratings_over_100 = df_ratings_movies.merge(df_high_value_rating, how='left')
df_ratings_over_100
```

```python
df_ratings_min_max_ts = df_ratings_over_100[df_ratings_over_100['rating_over_100'] == True].groupby('userId').agg({'timestamp':['min', 'max']}).reset_index()
df_ratings_min_max_ts.head()
```

```python
df_ratings_min_max_ts['diff'] = df_ratings_min_max_ts['timestamp']['max'] - df_ratings_min_max_ts['timestamp']['min']
```

```python
print(np.round(df_ratings_min_max_ts['diff'].mean(),2),'cекунд')
print('или')
print(np.round(df_ratings_min_max_ts['diff'].mean()/60,2),'минут')
print('или')
print(np.round(df_ratings_min_max_ts['diff'].mean()/3600,2),'часа')
print('или')
print(np.round(df_ratings_min_max_ts['diff'].mean()/(3600*24),2),'дня')
```
